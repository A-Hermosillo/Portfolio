### Aleatorización de orden

No queremos que el orden de nuestros datos afecte el entrenamiento de los modelos.

```python
X = df.iloc[: , 1:]
y = df.iloc[: , 0]
```

## Evaluación de Modelos

```python
kf = StratifiedKFold(n_splits = 10, shuffle = True)  
```

### SVC

#### Linear

```python
y_test_l = []
y_pred_l = []
accuracy  = 0 
cont = 1

for train_index, test_index in kf.split(X, y):
    # Training phase
    X_train = X.iloc[train_index, :]
    y_train = y.iloc[train_index]
    clf_SVCL = SVC(kernel = 'linear')
    clf_SVCL.fit(X_train, y_train)
    # Test phase
    X_test = X.iloc[test_index, :]
    y_test = y.iloc[test_index]    
    
    y_pred = clf_SVCL.predict(X_test)
    
    y_test_l.extend(y_test)
    y_pred_l.extend(y_pred)
    
    print('Exactitud en Fold', cont)
    print(accuracy_score(y_test, y_pred))
    print('-'*25)
    accuracy += accuracy_score(y_test, y_pred)
    cont += 1

print('Precisión por Clase')    
print(precision_score(y_test_l, y_pred_l, average = None))
print('Recall por Clase') 
print(recall_score(y_test_l, y_pred_l, average = None))
print('Exactitud promedio del modelo ')
accSVCL = accuracy/10
print(accSVCL)
```

## Obtención de Hiperparámetros

Cada uno de los modelos utilizados anteriormente utiliza hiperparámetros. A continuación se mencionarán los distintos tipos de modelos que se usaron y sus hiperparámetros:

**1. Support Vector Classifier**

Se utilizó este tipo de modelo dos veces durante la etapa de evaluación. Se usó dos veces precisamente por uno de sus hiperparámetros más importantes, el cual es el tipo de **kernel** que se va a utilizar. En este proyecto, se utilizaron un kernel de base radial y otro lineal, para ver cual sería más eficiente aunque hay varios tipos de kernels aparte de estos. Otro parámetro crítico es la penalización **C** que puede tomar un rango de valores y tiene un efecto dramático en la forma de las regiones resultantes para cada clase.

**2. K-Nearest Neighbors Classifier**

El hiperparámetro más importante de este modelo de clasificación es la cantidad de vecinos **n_neighbors** que se usaran. Otros hiperparámetros no tan significativos, pero aún así interesantes  incluyen a la contribución de los miembros del vecindario a través de diferentes ponderaciones **weights** y las distintas métricas de distancia para elegir la composición del vecindario **metric**.

**3. Multi-Layer Perceptron Classifier**

Un Perceptrón multicapa, al igual que la mayoría de las redes neuronales tiene varios hiperparámetros. Los más importantes son el número de capas ocultas y el número de unidades para cada capa oculta **hidden_layer_sizes**, la tasa de aprendizaje **learning_rate** y la función de activación para la capa oculta **activation**.


**4. Logistic Regression**

En este tipo de clasificador, realmente no hay hiperparámetros críticos para ajustar. Sin embargo, algunos hiperparámetros que pueden llegar a presentar cambios útiles en la exactitud son **solver**, el cual determina el algoritmo a utlizar en el problema de optimización, **penalty**, que dictamina el tipo de castigo a aplicar  y el parámetro **C** que controla la fuerza de la penalización.

**5. Random Forest Classifier**

El parámetro más importante es el número de características aleatorias para muestrear en cada punto de división **max_features**. Otro parámetro importante para el este tipo de clasificador es el número de árboles **n_estimators**. Idealmente, este parámetro debería incrementarse hasta que no se vea ninguna mejora adicional en el modelo. En este proyecto, el default de dicho parámetro fue suficiente para encontrar un aproximado del máximo rendimiento del modelo.

**¿Es posible entrenar un modelo de clasificación con los datos de un sujeto y obtener buenos resultados con los datos de otro sujeto? ¿Por qué?**

Si es posible. Esto se debe a que si uno entrena el modelo de clasificación con datos de un sujeto que realice las actividades de una manera parecida a como las realiza el otro sujeto, el modelo de clasificación si tendría buenos resultados. Sin embargo, como se ve en este ejercicio, si se tienen sujetos de distintas edades, géneros y condiciones físicas es muy complicado que el modelo de buenos resultados.

**¿Consideras que debes hacer ajustes cuando tengas datos de otro sujeto?**

Idealmente, al entrenar un modelo de clasificación se busca que la gente que ayude a proporcionar los datos sea una muestra poblacional para que el modelo funcione de manera global, con distintos tipos de sujetos. Este proyecto así se realizó y a lo mejor el único ajuste necesario puede ser un escalamiento de los datos. El inconveniente de esto es que la distribución de los datos puede presentar diferencias porque proviene de distintos sujetos. Otra opción, es entrenar distintos modelos para cada tipo de sujeto, pero esto toma más tiempo, se tienen menos datos y es más impráctico.

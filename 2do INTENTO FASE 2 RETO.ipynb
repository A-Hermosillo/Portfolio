```python
# GRAFICAS

# set width of bars
barWidth = 0.25
 
# set heights of bars
bars1 = [39710, 22934]
bars2 = [33434, 5357]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

# Make the plot
plt.bar(r1, bars1, color='#7400ff', width=barWidth, edgecolor='black', label='Hombres')
plt.bar(r2, bars2, color='#ffc900', width=barWidth, edgecolor='black', label='Mujeres')
 
# Add xticks on the middle of the group bars
plt.xlabel('Cantidad de muertes', fontweight='bold', color = 'black')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Obesidad', 'Tabaquismo'], color ='black')
plt.ylabel('Personas', fontweight='bold', color = 'black')
#plt.yticks(color = 'white')
# Create legend & Show graphic
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(range(10))
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.spines['bottom'].set_color('purple')
ax.spines['top'].set_color('purple')
ax.spines['left'].set_color('purple')
ax.spines['right'].set_color('purple')
ax.xaxis.label.set_color('red')
ax.tick_params(axis='x', colors='red')

plt.show()
```
